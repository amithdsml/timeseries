```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
tf_decompose = seasonal_decompose(wsb_df['Sale Quantity'],model='multiplicative',period=12)
```

```python
ts_plot = tf_decompose.plot()
```

```python
wsb_df['seasonal'] = tf_decompose.seasonal
wsb_df['trend'] = tf_decompose.trend
```
