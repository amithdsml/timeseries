```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
df=pd.read_csv('daily-total-female-births.csv')
df.head()
```

```python
df = pd.read_csv('daily-total-female-births.csv',parse_dates=['Date'],index_col=['Date'])
```

```python
df.head()
```

```python
from pylab import rcParams
rcParams['figure.figsize'] = 12,5
```

```python
df.plot()
```

```python
df_ad_decomp = seasonal_decompose(df,model='additive')
```

```python
df_ad_decomp.plot()
plt.show()
```

```python
df_ad_decomp.trend
```

```python
df_ad_decomp.seasonal
```

```python
df_ad_decomp.resid
```

```python
df_mul_decomp = seasonal_decompose(df,model='multiplicative')
df_mul_decomp.plot()
```

```python
df_mul_decomp.resid.plot()
```

```python
df_mul_decomp.seasonal.plot()
```
